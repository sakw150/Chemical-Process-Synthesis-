```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def tp(T_m):
    e_m=1
    sigma=5.67*10**-8
    a=0.07297
    T_W=1723.15
    u_g=3.2*10**-5
    V_m=0.365
    p_m=32.10
    delth=7.7756*10**4
    MW_m=0.04011
    return (e_m*sigma*a*(T_W**4-T_m**4))/u_g+(V_m*p_m*(-delth))/MW_m
```

```python
def euler(func,y0,tf,delt):
    n = int(tf/delt)
    t = np.zeros(n+1)
    y = np.zeros(n+1)
    t[0] = 0
    y[0] = y0
    for i in range(0,n):
        t[i+1] = t[i] + delt
        y[i+1] = y[i] + delt*func(t[i])
    plt.plot(t,y)
    plt.xlabel("Volume (L)")
    plt.ylabel("Temperature of Particle (K)")
    plt.title("Volume and Temperature of Particle")
    return y[i+1]
```

```python
euler(tp,298,2,0.05)
```
